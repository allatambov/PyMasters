### Задача 1

В переменной `user` сохранено имя пользователя (создайте переменную и сохраните любое). 

Напишите программу, которая запрашивает имя пользователя с клавиатуры и сравнивает его со значением в `user`. 

* Если они совпадают, то на экран выводится сообщение "Welcome back! Python missed you!".

* Если они не совпадают, то на экран выводится сообщение "Sorry, Python is busy and not available now".

### Задача 2

В списке `grades` сохранены оценки студентов за контрольную работу в 10-балльной шкале. 

    grades = [6, 7, 4, 3, 5, 10, 9, 8]
    
Напишите программу, которая перебирает все оценки в списке и на каждую оценку выводит текстовый комментарий ("Отлично", "Хорошо", "Удовлетворительно", "Плохо") и оценку в 5-ной шкале.

*Пример комментария (к оценке 7):*

    Хорошо, 4

*Пример комментария (к оценке 3):*

    Плохо, 2

### Задача 3

В списке `marks` сохранены пары имя-оценка в 5-балльной шкале:

    marks = [("Alicia", 2), ("Bill", 5), ("Charlie", 5), 
            ("Fred", 4), ("George", 4), ("Lily", 3), ("Susan", 3)]

Напишите программу, которая сохраняет *имена* студентов, получивших хорошие и отличные оценки, в список `good`, а получивших удовлетворительные и плохие оценки – в список `bad`.

**Подсказка:**

Один элемент списка – кортеж из двух элементов. Запустите следующий код и подумайте, как с таким списком можно работать:

    for pair in marks:
        print(pair[0])

### Задача 4

Напишите функцию `convert()`, которая принимает на вход целое число – рост человека в сантиметрах, а возвращает дробное число – рост человека в метрах. Функция не должна ничего печатать, то есть выводить на экран.

### Задача 5

Напишите функцию `money()`, которая принимает на вход сумму в рублях и курс евро и:

* возвращает сумму в евро, округленную до второго знака после запятой;
* выводит на экран сообщение вида "Сумма в рублях: [n1], сумма в евро: [n2]".

**Пример:**

*Применение:*

    money(1000, 70.77)

*Результат:*

    Сумма в рублях: 1000, сумма в евро: 14.13
    14.13

### Задача 6

Напишите функцию `dummy()`, которая принимает на вход список значений индекса Freedom House `f` и возвращает список из 0 и 1, где 1 соответствует свободным странам (статус *Free*, значения индекса от 1.0 до 2.5 включительно). Функция не должна ничего печатать, то есть выводить на экран.

**Примеры работы функции:**

*Применение:*

    dummy([2.5, 3.0, 2.7, 3.6, 4.8, 6.5, 1.0, 1.5, 7.0])

*Результат:*

    [1, 0, 0, 0, 0, 0, 1, 1, 0]

*Применение:*

    dummy([3.5, 3.0, 2.7, 3.6])

*Результат:*

    [0, 0, 0, 0]
